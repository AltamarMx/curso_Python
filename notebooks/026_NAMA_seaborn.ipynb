```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
nombres = np.genfromtxt("../data/nama.csv",max_rows=1,delimiter=",",
                       dtype="20U")
nombres
for i,nombre in enumerate(nombres):
    print(i,nombre)

nombres[4] = "reduccion_emisiones"

nombres
```

```python
archivo = '../data/nama.csv'
nama = pd.read_csv(archivo,names=nombres,skiprows=1)
# nama
```

```python
nama
```

```python
sns.displot(nama[nama.reduccion_emisiones>1],x="estado",hue="clima",
           aspect=3)
plt.xticks(rotation=90);
```

```python
sns.displot(nama[nama.reduccion_emisiones>1],x="municipio",hue="clima",
           aspect=10)
plt.xticks(rotation=90);
```

```python
sns.displot(nama[nama.reduccion_emisiones>1],x="municipio",hue="clima",
           aspect=3,row="estado")
plt.xticks(rotation=90);
```

```python
estados = nama.estado.unique()
```

```python

for estado in estados[:10]:
    sns.displot(nama[nama.estado==estado],x="municipio",hue="clima")
    plt.xticks(rotation=90);
    plt.title(estado)
```
