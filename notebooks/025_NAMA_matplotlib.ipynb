```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
nombres = np.genfromtxt("../data/nama.csv",max_rows=1,delimiter=",",
                       dtype="20U")
nombres
for i,nombre in enumerate(nombres):
    print(i,nombre)

nombres[4] = "reduccion_emisiones"

nombres
```

```python
archivo = '../data/nama.csv'
nama = pd.read_csv(archivo,names=nombres,skiprows=1)
nama
```

```python
nama.columns
```

```python
estados = nama['estado'].unique()
estados
```

```python
climas = nama["clima"].unique()
climas
```

```python
municipios = nama.municipio.unique()
municipios
```

```python
nama[nama.clima=="Seco y semiseco"]
```

```python
fig, ax = plt.subplots()

ax.set_xlim(0,500)
ax.hist(nama[nama.clima=="Seco y semiseco"].reduccion_emisiones,bins=200);
```

```python
fig, ax = plt.subplots(5,figsize=(10,8))

# ax.set_xlim(0,500)
for i,clima in enumerate(climas):
    
    ax[i].set_title(clima)
    ax[i].hist(nama[nama.clima==clima].reduccion_emisiones,
            bins=200);
fig.tight_layout()
```

```python
fig, ax = plt.subplots(33,figsize=(10,40))

# ax.set_xlim(0,500)
for i,estado in enumerate(estados):
    
    ax[i].set_title(estado)
    ax[i].hist(nama[nama.estado==estado].reduccion_emisiones,
            bins=200);
fig.tight_layout()
```
