```python
def sigmoide(x):
    return 1./(1. + np.exp(-x))
```

```python
x = np.linspace(-10,10,5000)
y = sigmoide(x)
plt.plot(x,y)
plt.grid()
```

```python
Y = 0.71
x = [3,2]
w = [0.1,0.2]
LR = 1
epochs = 50

def problema(x,w):
    return np.dot(x,w)

def sigmoide(x):
    return 1/(1 + np.exp(-x))

def costo(y,Y):
    return (y-Y)**2

def dCdw(y,Y):
    return 2*(y-Y)*y*(1-y)

y_h = []
C_h = []
for epoch in range(epochs):
    y = sigmoide(problema(x,w))
    C = costo(y,Y)
    correccion = dCdw(y,Y)*LR
    w = w - correccion
    C_h.append(C)
    y_h.append(y)
    print(epoch,C,y)
```

```python
plt.plot(y_h)
```

```python
plt.plot(C_h)
```

```python
def maquinita(x,w,Y,LR,epochs):
    y_h = []
    C_h = []
    for epoch in range(epochs):
        y = sigmoide(problema(x,w))
        C = costo(y,Y)
        correccion = dCdw(y,Y)*LR
        w = w - correccion
        C_h.append(C)
        y_h.append(y)
#         print(epoch,C,y)
    print(y)
    grafiquita(C_h,"r","Costo")
    grafiquita(y_h,"b","y")
    
    
    

def grafiquita(y,marca,ejey):
  plt.figure(figsize=(10,4))

  plt.plot(y,marca+"*")
  plt.plot(y,marca+"-")
  plt.ylabel(ejey)
  plt.xlabel("Epoch")
  plt.show()

    
```

```python
maquinita(x=[3,2],w=[0,0],Y=-1,LR=1,epochs=50)
```
