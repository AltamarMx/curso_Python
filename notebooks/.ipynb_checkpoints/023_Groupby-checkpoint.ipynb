# Heatmap horario de la temperatura promedio


```python
data.To.groupby(by=[data.index.month,data.index.hour]).mean().unstack().T
```

```python
fig, ax = plt.subplots(figsize=(10,4))
datos = data.To.groupby(by=[data.index.month,data.index.hour]).mean().unstack().T

ax.set_ylim(0,23)
ax.imshow(datos,cmap="jet",aspect=0.2)
```

```python
fig, ax = plt.subplots(figsize=(10,4))
datos = data.To.groupby(by=[data.index.month,data.index.hour]).mean().unstack().T

ax.matshow(datos,cmap="jet",aspect=0.2)
```
