```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "../data/Temixco_2018_patched_corrected_may2019_V1p0.csv"
data = pd.read_csv(file,index_col=0,parse_dates=True)
data
```

```python
data.Ig.groupby(by=[data.index.month,data.index.hour])
```

```python
data["Ig"].groupby(by=[data.index.month,data.index.hour]).mean().unstack().T
```

```python
fig, ax = plt.subplots(figsize=(10,10))


ax.imshow(data["Ig"].groupby(by=[data.index.month,data.index.hour]).mean().unstack().T)
```

```python
data["Ig"].groupby(by=[data.index.month,data.index.hour]).agg(["mean","std","max","min"])
```

```python
data["Ig"].groupby(by=[data.index.month,data.index.hour]).aggregate(["mean","std","max","min"]).plot(figsize=(10,4))
```

```python
data.Ig.groupby(by=[data.index.strftime("%H:%M")]).mean()
```
