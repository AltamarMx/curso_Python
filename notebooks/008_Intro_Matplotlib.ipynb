```python
# pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(2,10)
y = np.arange(2,18,2) 
y2 = np.arange(4,20,2) 
print(len(y), len(x))
```

```python
plt.plot(x,y,"r*",label="1")
plt.plot(x,y2,"k-*",label="2")

plt.legend()
plt.grid()
```

```python
x = np.arange(2,10)
y3 = 3*x**2
```

```python
y3
```

```python
plt.plot(x,y,"r*",  label="1")
plt.plot(x,y2,"k-*",label="2")
plt.plot(x,y3,"b.", label="3")

plt.legend()
plt.grid()
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.plot(x,y, "r+")
ax.plot(x,y2,"b+")
ax.plot(x,y3,"k-")

ax.set_ylim(-10,150)
ax.grid()
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,4),sharex=True)

ax[0].plot(x,y)
ax[1].plot(x,y3)

ax[0].set_ylim(0,20)
ax[1].set_ylim(0,200)
```

```python
fig, ax = plt.subplots(4,4,figsize=(10,10),sharex=True,sharey=True)
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,3))

ax[0].plot(x,y)

ax[1].plot(x,y3)
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,3))

ax[0][0].plot(x,y)

ax[0][1].plot(x,y3)
```

```python
y3 = 100*x**5
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(x,y ,"ko")
ax.set_ylabel("$y$",color="k")
ax.tick_params(axis='y', labelcolor="k")



ax2 = ax.twinx()
ax2.plot(x,y3,"r:")
ax2.set_ylabel("$y^5$",color="r")
ax2.tick_params(axis='y', labelcolor="r")


ax.set_xlabel("conteo [-]")
plt.savefig("figurita_200dpi.png",dpi=200)
plt.savefig("figurita_72dpi.png",dpi=72)
```
