subplot2grid(shape, loc)

shape: forma (tamaño [vertical, horizontal] del grid donde vamos a dibujar

loc: posicion donde empieza el dibujo [abajo, derecha]

```python
fig = plt.figure(figsize=(8,8))
ax0 = plt.subplot2grid((3,3), loc=(0,0))
ax0.annotate("ax0", (0.5,0.5))

ax1 = plt.subplot2grid((3,3), loc=(0,1))
ax1.scatter(tmx.To, tmx.RH)
ax1.set_xlabel("Humedad")
ax1.set_title("Plot2")

ax2 = plt.subplot2grid((3,3), loc=(1,0), colspan=2)
ax2.plot(tmx.RH)

ax3 = plt.subplot2grid((3,3), loc=(2,0), colspan=3)
ax3.annotate("ax3", (0.5,0.5))

ax4= plt.subplot2grid((3,3), loc=(0,2), rowspan=2)
ax4.annotate("ax4", (0.5,0.5))
ax4.set_ylabel("Radiacion")
ax4.set_title("Grafica4")
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(6,6))

ax0 = plt.subplot2grid(shape=(3,3), loc = (0,0), rowspan=2, colspan=2)
#ax0.scatter(tmx.RH, tmx.To, alpha=0.4)
ax0.hist2d(tmx["2018-12"].RH, tmx["2018-12"].To, bins=20)
#ax0.hexbin(tmx.RH, tmx.To, gridsize=20, alpha=0.1)

ax0.set_xlabel("Humedad")
ax0.set_ylabel("Temperatura")

ax1 = plt.subplot2grid(shape=(3,3), loc = (0,2), rowspan=2)
ax1.hist(tmx["2018-12"].To, density=True, bins=20, orientation="horizontal")
ax1.set_ylabel("Temperatura")

ax2 = plt.subplot2grid(shape=(3,3), loc = (2,0), colspan=2)
ax2.hist(tmx["2018-12"].RH, density=True, bins=20)
ax2.set_xlabel("Humedad")
plt.tight_layout()
```

```python
def fecha(mes):
    meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
            "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]
    fig = plt.figure(figsize=(6,6))

    ax0 = plt.subplot2grid(shape=(3,3), loc = (0,0), rowspan=2, colspan=2)
    #ax0.scatter(tmx.RH, tmx.To, alpha=0.4)
    ax0.hist2d(tmx[f"2018-{mes}"].RH, tmx[f"2018-{mes}"].To, bins=20)
    #ax0.hexbin(tmx.RH, tmx.To, gridsize=20, alpha=0.1)

    ax0.set_xlabel("Humedad")
    ax0.set_ylabel("Temperatura")

    ax1 = plt.subplot2grid(shape=(3,3), loc = (0,2), rowspan=2)
    ax1.hist(tmx[f"2018-{mes}"].To, density=True, bins=20, orientation="horizontal")
    ax1.set_ylabel("Temperatura")

    ax2 = plt.subplot2grid(shape=(3,3), loc = (2,0), colspan=2)
    ax2.hist(tmx[f"2018-{mes}"].RH, density=True, bins=20)
    ax2.set_xlabel("Humedad")
    
    plt.suptitle(meses[mes-1])
    plt.tight_layout()
    
    plt.savefig(f"./histogramas/Histograma_{mes:02}.png")
    
    print(f"Listo imagen {i}")
```
