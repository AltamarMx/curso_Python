# 1. El atractor de Lorenz

$$ \begin{align}
& \frac{dx}{dt}=\sigma (y-x) \\
& \frac{dy}{dt}=x(\rho-z)-y \\
&\frac{dz}{dt}=xy-\beta z 
\end{align}$$


```python
import numpy as np
import matplotlib.pyplot as plt
from celluloid import Camera #pip install Celluloid
```

```python
def lorenz(x, y, z, s=15, r=20, b=2.667):
    x_dot = s*(y-x)
    y_dot = r*x-x*z-y
    z_dot = x*y-b*z
    return x_dot, y_dot, z_dot
```

```python
dt = 0.01
num_steps = 10000
```

```python
xs = np.empty(num_steps + 1)
ys = np.empty(num_steps + 1)
zs = np.empty(num_steps + 1)
```

```python
xs[0], ys[0], zs[0] = (0., 1.5, 1.)
```

```python
for i in range(num_steps):
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i+1] = xs[i] + (x_dot*dt)
    ys[i+1] = ys[i] + (y_dot*dt)
    zs[i+1] = zs[i] + (z_dot*dt)
    
```

```python
plt.plot(xs, ys)
```

```python
plt.plot(zs, ys)
```

```python
ax = plt.figure().add_subplot(projection="3d")

ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
ax.set_title("Atractor de Lorenz")

```

## 1.1 Gif a partir de un conjunto de imágenes

```python
dt = 0.01
num_steps = 5000

xs = np.empty(num_steps + 1)
ys = np.empty(num_steps + 1)
zs = np.empty(num_steps + 1)

xs[0], ys[0], zs[0] = (0., 1., 1.)

#ax = plt.figure().add_subplot(projection='3d')
for i in range(num_steps):
    print("empezar")
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + (x_dot * dt)
    ys[i + 1] = ys[i] + (y_dot * dt)
    zs[i + 1] = zs[i] + (z_dot * dt)
    
    

    if i%50 == 0:
        
        ax = plt.figure().add_subplot(projection='3d')
        ax.plot(xs[:i], ys[:i], zs[:i], lw=0.5, color='k');
        ax.scatter(xs[i], ys[i], zs[i], color='red')
        ax.set_xlabel("X Axis");
        ax.set_ylabel("Y Axis");
        ax.set_zlabel("Z Axis");
        ax.set_title("Lorenz Attractor")
        ax.set_xlim(-20,20) #min(x), min(y)
        ax.set_ylim(-20,30)
        ax.set_zlim(0,50)
        plt.savefig(f"./img/lorenz_{i:04}.png")
        plt.close()
    print("Terminé")
```

```python
%%time
import numpy as np
import matplotlib.pyplot as plt
from celluloid import Camera
# https://www.askpython.com/python-modules/animation-in-python-celluloid

def lorenz(x, y, z, s=10, r=28, b=2.667):
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot


dt = 0.01
num_steps = 5000

# Need one more for the initial values
xs = np.empty(num_steps + 1)
ys = np.empty(num_steps + 1)
zs = np.empty(num_steps + 1)

xs[0], ys[0], zs[0] = (1., 1.9, 1.)

# Step through "time", calculating the partial derivatives at the current point
# and using them to estimate the next point


fig = plt.figure()
ax = fig.gca(projection='3d')
camera = Camera(fig)

for i in range(num_steps):
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + (x_dot * dt)
    ys[i + 1] = ys[i] + (y_dot * dt)
    zs[i + 1] = zs[i] + (z_dot * dt)

    if i%50 == 0:
        print(i)
        ax.plot(xs[:i], ys[:i], zs[:i], lw=0.5, color='k');
        ax.scatter(xs[i], ys[i], zs[i], color='red')
        ax.set_xlabel("X Axis");
        ax.set_ylabel("Y Axis");
        ax.set_zlabel("Z Axis");
        ax.set_title("Lorenz Attractor");
        camera.snap()
        #plt.close()
    
animation = camera.animate()
#animation.save("lorenz_2.gif", writer="imagemagick")
animation.save("lorenz_celluloid.gif")
```

## 2. Grafica de una superficie


```python
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib.animation as animation
```

```python
t = np.linspace(0,10,100)
x = np.linspace(-np.pi, np.pi, 11)
y = np.linspace(-np.pi, np.pi, 11)

xmesh, ymesh = np.meshgrid(x,y)
```

```python
def funcion(x,y,t):
    return np.sin(y+t)*np.cos(x)
```

```python
fig = plt.figure()
ax  = fig.gca(projection="3d")
```

```python
def actualizar(i):
    ax.clear()
    zmesh = funcion(xmesh, ymesh, t[i])
    ax.plot_surface(xmesh, ymesh,zmesh, cmap="viridis")
#     plt.xlim(min(x), max(x))
#     plt.ylim(min(y), max(y))
```

```python
ani = animation.FuncAnimation(fig, actualizar, range(len(t)),interval=10, repeat=False)
```

```python
ani.save("superficie.gif")
```
