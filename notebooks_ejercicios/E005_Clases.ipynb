## 2. Con cada una de las funciones previamente definidas, crea una clase llamada Viento.

````python
class Viento:
    def __init__(self, vels):
        self.vels = vels
        
    def promedio(self):
        return sum(self.vels)/len(self.vels)

    def cuartiles(self, k):
        self.vels.sort()

        #caso impar
        if len(self.vels)%2 != 0:  #!= no es igual
            Q = k*(len(self.vels)+1)//4 
            return(self.vels[(Q-1)])

        #caso par
        #elif len(vels)%2 == 0  
        else:
            Q = k*(len(self.vels))//4
            return((self.vels[Q]+self.vels[Q-1])/2)
        
    def potencias(self, rho=1.225, A=1):

        pots = [0.5*rho*datos**3*A for datos in self.vels]
    
        return pots
````

````python
campeche = Viento(velocidades)
````

````python
campeche.promedio()
````

````python
campeche.cuartiles(3)
````

````python
campeche.potencias(A=2)
````

## 3. Importa los datos del archivo temixco_2018
Usa los datos de la columna WS, conviertelos a una lista y aplica los métodos de Viento

````python
file = open("./data/temixco_2018.csv") #abrir el archivo
lines = file.readlines()    #leer línea por línea, todas las líneas
file.close()                #cerrar el archivo

result = []                 #una lista nueva donde guardaremos los datos
for x in lines:             #iterar sobre las líneas
    result.append(x.rstrip("\n").split(",")[1])   #eliminar espacios vacíos, dividir por comar, append los elementos 
                                                #de la columna 2
    
# pasar todos los elementos a flotante
velocidades = []            
for dato in result[1:]:
    velocidades.append(float(dato))
````

````python
temixco = Viento(velocidades)
````
