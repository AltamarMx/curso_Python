```python
def importa_sin(path):
    nombres = ["Dia","Hora","Eolica","Fotovoltaica","Biomasa","Carboelectrica","CicloCombinado",
           "CombustionInterna","Geotermoelectrica","Hidroelectrica","Nucleoelectrica",
           "TermicaConvencional","TurboGas"]
    data = pd.read_csv(path, skiprows=8,
                  usecols=np.arange(1,14,1), names=nombres)
    data.Hora -= 1
    data["Tiempo"] = data["Dia"] + data["Hora"].astype(str)
    data.Tiempo = pd.to_datetime(data.Tiempo, format="%d/%m/%Y%H")
    data.set_index("Tiempo", inplace=True)
    data.drop(["Dia", "Hora"], axis=1, inplace=True)
    return data
```

## Concat multiples archivos

```python
#Nota: Por la naturaleza de los archivos es posible crear un try, except para poder "ignorar" los 
#archivos que tengan el '25' como marca temporal
def import_sin(path):
    data_1 = pd.read_csv(path,
                    skiprows=8, usecols=np.arange(1,14,1), names=nombres)
    data_1.Hora -= 1
    data_1["Tiempo"] = data_1["Dia"] + data_1["Hora"].astype(str)
    try:
        data_1.Tiempo = pd.to_datetime(data_1.Tiempo, format="%d/%m/%Y%H")
        data_1.set_index("Tiempo", inplace=True)
        data_1.drop(["Dia", "Hora"], axis=1, inplace=True)
        return data_1
    except ValueError:
        print("Dataframe sin formato " + path)
    
```

```python
import glob

#Crea una lista con todos los archivos en esa carpeta
files = glob.glob("./data/SIN/2019/Generacion Liquidada_L2 SEN *.csv")
```

```python
todos = [] #lista para guardar todos los df

#itera sobre la lista de paths e importa cada uno de los archivos
for i in files:
    todos.append(import_sin(i))
```

```python
#concatenar
todos = pd.concat(todos)
```

```python
#ordenar
todos.sort_index(ascending=True)
```
